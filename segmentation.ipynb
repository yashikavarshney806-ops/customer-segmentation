```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education'])
df['Marital_Status'] = le.fit_transform(df['Marital_Status'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scld=scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(scld)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("No. of clusters")
plt.ylabel("WCSS")
plt.show
```

```python
# validating value of k
from kneed import KneeLocator
```

```python
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
```

```python
kmeans_final=KMeans(n_clusters=kl.elbow,init='k-means++')
clusters=kmeans_final.fit_predict(scld)
```

```python
from sklearn.metrics import silhouette_score
silhouette_coeff=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=  k,init='k-means++')
    kmeans.fit(scld)
    score=silhouette_score(scld,kmeans.labels_)
    silhouette_coeff.append(score)
silhouette_coeff
```

```python
plt.plot(range(2,11),silhouette_coeff)
plt.xlabel("No. of clusters")
plt.ylabel("Silhouette_coeff")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_data=pca.fit_transform(scld)
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1],c=clusters,cmap='plasma')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
print(pca.explained_variance_ratio_)
```
